```python

lexicolist = []
token = ''
classe = ''
linha = ''
vez = 1
flag = 0
listrange = 0

class lex:  
    def __init__(self, token, classe, linha):  
        self.token = token  
        self.classe = classe 
        self.linha = linha 


for line in tabela: 
    
    for word in line.split():
        #print(word)
        flag = 0
        if (vez == 1) and (flag == 0):
            token = word 
            flag = 1
            vez = 2
            #print(token)
        
        if (vez == 2) and (flag == 0):
            classe = word
            flag = 1
            vez = 3
            #print(classe)
        
        if (vez == 3) and (flag == 0):
            linha = word
            flag = 1
            vez = 1
            #print(linha)
    
    lexicolist.append(lex(token, classe, linha))  
    listrange = listrange + 1
    #print (lexicolist)

for obj in lexicolist: 
    print( obj.token, obj.classe, obj.linha, sep = ' ' )
    


#lexico = iter(())

#lexico = iter(lexico)






```

```python
lexposition = 0
okflag = True

def nextlexico():
    global lexposition
    global okflag
    if lexposition != (listrange - 1):  
        lexposition = lexposition + 1    
        okflag = True
        return lexicolist[lexposition]
    else:
        print('out of list range')
        return lexicolist[lexposition]
def backlexico(x):
    global lexposition
    lexposition = lexposition - x
    okflag = False
    return lexicolist[lexposition]
    
```

```python

#lexico = iter(lexicolist)
lexposition = 0
okflag = True

def errormsg(x,line):
    x = x
    line=line
    print('Error: a "',x,'" was expected on line', line)
    
            

def ok():
    global x
    x = nextlexico()
    


#-----------------------------------------------------------------------------------------
#----------------------------------------------------------------------------------------


def analisador_sintatico():
    global x
    x = lexicolist[0]
    programa()
    if okflag == False:
        print('ERRO FATAL')
        
    
#---------------------------------------------------------------------
    
def programa():
    global x
    #print(x.token)
    if x.token == 'program':
        ok()
        if x.classe == 'Identificador':
            ok()
            ok()
            print('inicio declaracoes_variaveis')
            declaracoes_variaveis()
            print('fim declaracoes_variaveis')
            declaracoes_subprog()
            print('fim declaracoes_subprog')
            comando_composto()
            print('fim declaracoes_comando_composto')
            
            #x = (next(lexico))
            if (x.token != '.'):                
                errormsg('.', x.linha)
                backlexico(0)
            
        else:
            errormsg('identifier', x.linha)
    else: 
        errormsg('program declaration', x.linha)
    
#------------------------------------------------------------------------    
    
def declaracoes_variaveis():
    global x
    #x = (next(lexico))
    if x.token == 'var':
        ok()
        lista_declaracoes_var()                    
    
    else:
        print('variavel nao foi declarada', x.linha)
#------------------------------------------------------------------------    
    
    
def lista_declaracoes_var():
        global x
        
        lista_de_identificadores()
        #x = (next(lexico))
        if x.token == ':':
                ok()
                tipo()
                #x = (next(lexico))
                if x.token == ';':
                        lista_declaracoes_var_aux()
                else: 
                    errormsg(';', x.linha)
                    backlexico(1)
        else:
            errormsg(':', x.linha)
            backlexico(0)
            
#--------------------------------------------------------------------    
    
    
def lista_declaracoes_var_aux():
        global x    
        lista_de_identificadores()
        #x = (next(lexico))
        if x.token == ':':
                ok()
                tipo()
                #x = (next(lexico))
                if x.token == ';':
                        ok()
                        lista_declaracoes_var_aux()
                else: 
                    errormsg(';', x.linha)
                    backlexico(1)
        else:
            errormsg(':', x.linha)
            backlexico(0)
    
#-------------------------------------------------------------------------    
    
    
def lista_de_identificadores():
        global x
        #x = (next(lexico))
        if x.classe == 'Identificador':
                ok()
                lista_de_identificadores_aux()
        else:
            errormsg('identifier', x.linha)
            backlexico(0)
        
#----------------------------------------------------------------------------          
        
                
    
def lista_de_identificadores_aux():
        global x
        #x = (next(lexico))
        if x.token == ',':
                ok()
                if x.classe == 'Identificador':
                        ok()
                        lista_de_identificadores_aux()
                else:  
                        errormsg('identifier', x.linha)
                        backlexico(0)
        
#---------------------------------------------------------------------------------------   
    
    
def tipo():
    global x
    #x = (next(lexico))
    if x.token == 'integer':
        ok()
    elif x.token == 'real':    
        ok()
    elif x.token == 'boolean':
        ok()
    else:
        errormsg('type', x.linha)
        backlexico(0)
    
    

#----------------------------------------------------------------------------------    
    

def declaracoes_subprog():
        declaracoes_subprog_aux()
        
#----------------------------------------------------------------------------------------    
def declaracoes_subprog_aux():
        global x 
        declaracao_subprog()
        #x = (next(lexico))
        if x.token == ';':
                ok()
                declaracoes_subprog_aux()
        else:
            errormsg(';', x.linha)
            backlexico(0)
#----------------------------------------------------------------------------------------    

def declaracao_subprog():
        global x
        #x = (next(lexico))
        if x.token == 'procedure':
                ok()
                #x = (next(lexico))
                if x.classe == 'Identificador':
                        ok()
                        argumentos()
                        #x = (next(lexico))
                        if x.token == ';':
                            ok()
                            declaracoes_var()
                            declaracoes_subprog()
                            comando_composto()                        
                        
                        else:
                            errormsg(';', x.linha)
                            backlexico(2)
                else:
                    errormsg('identifier', x.linha)
                    backlexico(1)
        #else:
        #   errormsg('procedure declaration')
        

#---------------------------------------------------------------------------------------------        
        
def argumentos():
    print('inicio argumentos')
    global x
    #x = (next(lexico))
    if x.token == '(':
            ok()
            lista_de_parametros()
            #x = (next(lexico))
            if x.token != ')':
                    errormsg(')', x.linha)
                    backlexico(1)

        
#------------------------------------------------------------------------------------------

   
    
def lista_de_parametros():
    
    global x
    lista_de_identificadores()
    #x = (next(lexico))
    if x.token == ':':
        ok()
        tipo()
        lista_de_parametros_aux()
        
    else: 
        errormsg(":", x.linha)
        backlexico(0)
# 
#-----------------------------------------------------------------------        
        
def listad_de_parametros_aux():
    global x
    lista_de_identificadores()
    #x = (next(lexico))
    if x.token == ':':
        ok()
        tipo()
        lista_de_parametros_aux()
    else: 
        errormsg(":", x.linha)
        backlexico(0)

#----------------------------------------------------------------------------    
           
        
def comando_composto():
    global x
    #x = (next(lexico))
    if x.token == 'begin':
        ok()
        comandos_opcionais()
        #x = (next(lexico))
        if x.token == 'end':
            ok()
        else:
            errormsg('end', x.linha)
            backlexico(1)
    else:
            errormsg('begin', x.linha)         
            backlexico(0)
    
    
#-----------------------------------------------------------------------    
    
def comandos_opcionais() :
            lista_de_comandos()
    
    
    
#-------------------------------------------------------------------------


def lista_de_comandos():
            comando()
            lista_de_comandos_aux()
    

    


#-------------------------------------------------------------
def lista_de_comandos_aux():
            #x = (next(lexico))
            if x.token == ';':
                    ok()
                    comando()
                    lista_de_comandos_aux()
            else:
                errormsg(';', x.linha)
                backlexico(0)
                
                    
    
    
#lista_de_comandos_aux →  ; comando lista_de_comandos_aux
#                        |@
    


#---------------------------------------------------------------------------

def comando():
    global x
    if x.token == 'if':
        ok()
        expressao()
        if x.token == 'then':
            ok()
            comando()
            parte_else()        
        errormsg('then', x.linha)
    elif x.token == 'while':
        ok()
        expressao()
        if x.token == 'do':
            comando()
    if okflag == False:
        comando_composto()
    
    if okflag == False:
        variável()
        if okflag == True:
            if x.token == ':=':
                ok()
                expressao()
            else:
                backlexico(1)
                
        
        
            
        
#comando → variável := expressão
#        | ativação_de_procedimento 
#        | comando_composto
#        | if expressão then comando parte_else 
#       | while expressão do comando 

    
#--------------------------------------------------------------------------    


def parte_else():
    #x = (next(lexico))
    if x.token == 'else':
        ok()
        comando();
    else:
        backlexico(0)    
    

#------------------------------------
def variavel():
    global x
    #x = (next(lexico))
    if x.token == 'id':
        ok()
    else:
        backlexico(0)
        
    
#variável → id


#--------------------------------------

def ativacao_de_procedimento():
        if x.token == 'id':
            ok()
            ativacao_de_procedimento_aux() 
        else:
            backlexico(0)
#ativação_de_procedimento → id Ativação_de_procedimento_aux  

#------------------------------------------------

def ativacao_de_procedimento_aux():
        
        if x.token == '(':
                ok()
                lista_de_expressoes()
                if x.token == ')':
                    ok()
                else: 
                    errormsg(')', x.linha)
                    backlexico(1)
        else:
            errormsg('(', x.linha)
            backlexico(0)
    
#Ativação_de_procedimento_aux → (lista_de_expressões) 
#                            |@   
    
#-----------------------------------------------------------

def lista_de_expressoes():
    expressao()
    lista_de_expressoes_aux()
    

#ista_de_expressões → expressão  lista_de_expressões_aux

#-------------------------------------------------------------

def lista_de_expressoes_aux():
    if x.token == ',':
        ok()
        expressao()
        lista_de_expressoes_aux()
    
    
#lista_de_expressões_aux →, expressão lista_de_expressões_aux 
#				|@
#---------------------------------------------------------------------

def expressao():
    expressao_simples()
    expressao_aux()


def expressao_aux():
    op_relacional()
    expressao_simples()
    
    
#expressão → expressão_simples   expressao _aux



#expressao _aux →  op_relacional expressão_simples 
#			|@

#_--------------------------------------------------------------------

def expressao_simples():
    termo()   
    expressao_simples_aux()    
    
    if okflag == False:
            sinal()
            termo()
            expressao_simples_aux()
                       

def expressao_simples_aux():
    op_aditivo()
    termo()
    expressao_simples_aux()   
    
    
#expressão_simples → termo Expressão_simples_aux
#             | sinal termo Expressão_simples_aux


#Expressão_simples_aux → op_aditivo termo Expressã#o_simples_aux 
#	| @

#----------------------------------------------------------------------------


def termo():
    fator()
    if okflag == False:
        termo()
        op_multiplicativo()
        fator()


#termo → fator 
#| termo op_multiplicativo fator 





#------------------------------------------------------------------------------------------



def fator():
    
    if x.classe == 'Identificador':
        ok()
    
    elif x.classe == 'NumeroInteiro':
        ok()
    elif x.classe == 'NumerReal':
        ok()
    elif x.token == 'true':
        ok()
    elif x.token == 'false':
        ok()
    elif x.token == '(':
        ok()    
        expressao()
        if x.token ==')':
            ok()
            
    elif x.token == 'not':
        ok()
        fator()
        
    else: 
        backlexico(0)
    
    
        
def fator_aux():
    if x.token == '(':
        ok()    
        lista_de_expressoes()
        if x.token ==')':
            ok()
        else:
            errormsg(')', x.linha)
            backlexico(1)
    else:
        errormsg('()', x.linha)
        backlexico(0)
        
        


#fator → 
#id Fator_aux 
#| num_int 
#| num_real 
#| true 
#| false 
#| (expressão) 
#| not fator

#Fator_aux → ( lista_de_expressões ) 
#		|@

#------------------------------------------------------------

def sinal():
    if x.token == '+':
        ok()
    elif token == '-':
        ok()
    else: 
        backlexico(0)
        
def op_relacional():
    if x.token == '=':
        ok()
    elif token == '<':
        ok()
    elif token == '>':
        ok()
    elif token == '<=':
        ok()
    elif token == '>=':
        ok()
    elif token == '<>':
        ok()       
    
    else: 
        backlexico(0)
        
        
def op_aditivo():
    if x.token == '+':
        ok()
    elif token == '-':
        ok()
    elif  token == 'or':
        ok() 
    else: 
        backlexico(0)
    
def op_multiplicativo():
    
    if x.token == '*':
        ok()
    elif token == '/':
        ok()
    elif  token == 'and':
        ok() 
    else: 
        backlexico(0)
    

#sinal → + | -

#op_relacional → = | < | > | <= | >= | <>

#op_aditivo → + | - | or

#op_multiplicativo → * | / | and

    
analisador_sintatico()

tabela.close()
file.close()    
```
